```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from  sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset )
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
x_train.shape,y_train.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test) #accuracy is should be good if not then keep changing random state until 
#acuracy gets better
```

```python
test_value=pd.DataFrame({"cgpa":[6.89]})
```

```python
lr.predict(test_value) #same ans as y=mx+c 
```

```python
#y=mx+c
lr.coef_
```

```python
lr.intercept_
```

```python
0.57515379*6.89 -1.0099739352598416
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,3))
plt.title("LINEAR REGRESSION")

sns.scatterplot(x="cgpa",y="package",data=dataset , label="orginal_data")

plt.plot(dataset["cgpa"],y_prd,color="red" , label="predicted_data")

plt.xlabel("cgpa")
plt.ylabel("package")
plt.legend()
plt.savefig("placement.jpg")
plt.show()
```
