```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("st_placement.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(4,2))
sns.scatterplot(x="cgpa" , y="resume_score" ,data=dataset ,hue="placed")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train ,x_test ,y_train , y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

```python
gn=GaussianNB()
```

```python
gn.fit(x_train ,y_train)
```

```python
gn.score(x_test ,y_test) *100 ,gn.score(x_train ,y_train) *100
```

```python
gn.predict([[8.14	,6.52]])
```

```python
plt.figure(figsize=(7,3))
plot_decision_regions (x.to_numpy() ,y.to_numpy(), clf=gn)
plt.show()
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train ,y_train)
```

```python
mnb.score(x_test ,y_test) *100 ,mnb.score(x_train ,y_train) *100
```

```python
plt.figure(figsize=(7,3))
plot_decision_regions (x.to_numpy() ,y.to_numpy(), clf=mnb)
plt.show()
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_train ,y_train)
```

```python
bnb.score(x_test ,y_test) *100 ,bnb.score(x_train ,y_train) *100
```

```python
plt.figure(figsize=(7,3))
plot_decision_regions (x.to_numpy() ,y.to_numpy(), clf=bnb)
plt.show()
```
