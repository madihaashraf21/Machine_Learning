```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns="EstimatedSalary" , inplace =True)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age" ,y="Purchased" ,data=dataset)
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train ,y_train)
```

```python
lr.score(x_test , y_test) *100
```

```python
test_value=pd.DataFrame({"Age":[40]})
```

```python
lr.predict(test_value)
```

```python
test_size=lr.predict(x)
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="Age" ,y="Purchased" ,data=dataset)
sns.lineplot(x="Age", y=test_size  , data=dataset ,color="red")
plt.show()
```
